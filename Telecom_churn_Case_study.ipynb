### Apply PCA to the Dataframe

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(telecom_high_value_df_scaled)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

### Making screeplot of variance ratio

```python
variance_cumsum = np.cumsum(pca.explained_variance_ratio_)
```

```python
fig = plt.figure(figsize=(12,8))
plt.vlines(x=53, ymax=1, ymin=0, colors='r', linestyles='--')
plt.hlines(y=0.90, xmax=200, xmin=0, colors='g', linestyles='--')
plt.plot(variance_cumsum)
plt.title("Scree Plot")
plt.ylabel("Cumulative variance explained")
plt.show()
```

53 component explain 90% of variance 

```python
from sklearn.decomposition import IncrementalPCA
```

```python
final_pca = IncrementalPCA(n_components=53)
```

```python
X_train_pca = final_pca.fit_transform(X_train)
```

```python
X_train_pca.shape
```

```python
X_test_pca = final_pca.transform(X_test)
```

### Create logistic regresion model on principle components

```python
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
```

```python
X_train_sm = sm.add_constant(X_train_pca)
log_model1 = sm.GLM(y_train, X_train_sm, family=sm.families.Binomial()).fit()
log_model1.summary()
```

```python
learner_pca = LogisticRegression()
model_pca = learner_pca.fit(X_train_pca, y_train)
pred_probs_test = model_pca.predict_proba(X_test_pca)
```

```python
from sklearn import metrics
"{:2.2}".format(metrics.roc_auc_score(y_test, pred_probs_test[:,1]))
```

```python
y_test_pred = model_pca.predict(X_test_pca)
```

```python
from sklearn.metrics import confusion_matrix
con_mat = confusion_matrix(y_test, y_test_pred)
```

```python
print(con_mat)
```

```python
TP = con_mat[1,1]
FN = con_mat[1,0]
TN = con_mat[0,0]
FP = con_mat[0,1]
```

```python
# lets calculate the sensitivity = TP/(FN+TP)
print(round(TP/(TP+FN)*100,2))
```

```python
# calculate the specificity
print(round(TN/(FP+TN)*100 ,2))
```

```python
print(metrics.accuracy_score(y_test, y_test_pred) * 100)
```

```python
from sklearn.metrics import roc_auc_score

# check area under curve
y_pred_prob = model_pca.predict_proba(X_test_pca)[:, 1]
print(roc_auc_score(y_test, y_pred_prob))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Create the parameter grid based on the results of random search 
param_grid = {
    'max_depth': [4,8,10,15],
    'min_samples_leaf': range(100, 400, 200),
    'min_samples_split': range(200, 500, 200),
    'n_estimators': [100,200, 300], 
    'max_features': [5, 10, 15]
}
# Create a based model
rf = RandomForestClassifier()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1,verbose = 1)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
# printing the optimal accuracy score and hyperparameters
print('We can get accuracy of',grid_search.best_score_,'using',grid_search.best_params_)
```

```python
rfc = RandomForestClassifier(class_weight={0:0.1, 1: 0.9}, n_estimators=100,
                             max_features=15, max_depth=4,
                            min_samples_split=100)
```

```python
# fit
rfc.fit(X_train,y_train)
```

```python
# predict
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,predictions))
```

```python
confusion = confusion_matrix(y_test,predictions)
```

```python
print(accuracy_score(y_test,predictions))
```

```python
TP = confusion[1,1]
FN = confusion[1,0]
TN = confusion[0,0]
FP = confusion[0,1]
```

```python
# lets calculate the sensitivity = TP/(FN+TP)
print(round(TP/(TP+FN)*100,2))
```

```python
# calculate the specificity
print(round(TN/(FP+TN)*100 ,2))
```

After building number of models and tuning random forest hyper-paramenters. Got specifity of 98% and sensitivity of 71% so as compare to the roandom forest combination of PCA and logistic regression perform better
